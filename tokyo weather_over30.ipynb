```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read file
df = pd.read_csv("kion10y.csv", encoding="utf-8")
```

```python
# pull data with temp>30
atui = df[df["temp"]>30]
# count per year
cnt = atui.groupby(["year"])["year"].count()
```

```python
# output
print(cnt)
cnt.plot()
plt.savefig("day_over30.png")
plt.show()
```
